````python
# prompt: apply autoencoder on this

import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from sklearn.model_selection import train_test_split

# Select only numeric columns relevant to salary
features = ["Basic Salary", "Allowance", "Statutory Bonus", "Gross Salary",
            "Total Deductions", "Net Salary", "Tax Spend", "Reimbursement Paid", "Deductions %", "Salary/Experience"] # Include newly created features

X = data[features].copy() # Use the data with the new features

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and testing sets (optional, but good practice)
# Autoencoders are typically trained on "normal" data.
# For anomaly detection, you might train on a subset assumed to be normal,
# or train on the whole dataset and identify points with high reconstruction error.
# Here, we'll train on the whole scaled dataset.
X_train = X_scaled

# Define the Autoencoder model
input_dim = X_train.shape[1]
encoding_dim = int(input_dim / 2) # You can tune this

input_layer = Input(shape=(input_dim,))
encoder = Dense(encoding_dim, activation="relu")(input_layer)
decoder = Dense(input_dim, activation="linear")(encoder) # Linear activation for reconstruction

autoencoder = Model(inputs=input_layer, outputs=decoder)

# Compile the Autoencoder
autoencoder.compile(optimizer='adam', loss='mse') # Mean Squared Error is a common loss

# Train the Autoencoder
# Use a small batch size and more epochs for potentially better reconstruction
history = autoencoder.fit(X_train, X_train,
                          epochs=50, # You can adjust the number of epochs
                          batch_size=32, # You can adjust the batch size
                          shuffle=True,
                          validation_split=0.1, # Use a small validation split
                          verbose=1) # Set to 0 to suppress output during training

# Plot training and validation loss (optional)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()


# Get reconstruction errors
reconstructions = autoencoder.predict(X_scaled)
mse = np.mean(np.power(X_scaled - reconstructions, 2), axis=1)

# Set a threshold for anomaly detection
# This is a crucial step and often requires experimentation
# You can use methods like the mean + std, or visualize the error distribution
threshold = np.percentile(mse, 95) # Example: set threshold at the 95th percentile of errors
print(f"Anomaly threshold (MSE): {threshold}")


# Classify anomalies based on the threshold
data['Reconstruction_Error'] = mse
data['Anomaly_AE'] = data['Reconstruction_Error'] > threshold

# Map boolean result to labels
data['Anomaly_AE'] = data['Anomaly_AE'].map({True: "Anomaly", False: "Normal"})

print("\nAutoencoder Anomaly Detection Results:")
print(data['Anomaly_AE'].value_counts())

# Display anomalies (optional)
anomalies_ae = data[data['Anomaly_AE'] == 'Anomaly']
print("\nExamples of Anomalies detected by Autoencoder:")
print(anomalies_ae.head())

# Visualize the distribution of reconstruction errors
plt.figure(figsize=(10, 6))
sns.histplot(mse, bins=50, kde=True)
plt.axvline(threshold, color='r', linestyle='--', label=f'Threshold ({threshold:.4f})')
plt.title('Distribution of Reconstruction Errors')
plt.xlabel('Mean Squared Error (MSE)')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Compare with OneClassSVM results (optional)
print("\nComparison of Anomaly Flags:")
print(data[['Anomaly', 'Anomaly_AE']].value_counts())
````

````python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seeds for reproducibility
np.random.seed(42)
random.seed(42)

def generate_autoencoder_test_data(n_records=50, anomaly_rate=0.20):
    """
    Generate synthetic payroll test data specifically for evaluating autoencoder performance
    Focus on patterns that autoencoders should detect based on reconstruction error
    """

    # Business units with consistent patterns
    business_units = {
        'IT': {'base_salary': 75000, 'std': 15000, 'allowance_rate': 0.12, 'deduction_rate': 0.15},
        'HR': {'base_salary': 55000, 'std': 10000, 'allowance_rate': 0.10, 'deduction_rate': 0.13},
        'Finance': {'base_salary': 68000, 'std': 12000, 'allowance_rate': 0.11, 'deduction_rate': 0.16},
        'Marketing': {'base_salary': 62000, 'std': 11000, 'allowance_rate': 0.09, 'deduction_rate': 0.14},
        'Operations': {'base_salary': 48000, 'std': 8000, 'allowance_rate': 0.08, 'deduction_rate': 0.12},
        'Sales': {'base_salary': 58000, 'std': 13000, 'allowance_rate': 0.13, 'deduction_rate': 0.14}
    }

    # Employee levels with consistent multipliers
    levels = {
        'Junior': {'multiplier': 0.8, 'exp_range': (0, 2)},
        'Mid': {'multiplier': 1.0, 'exp_range': (2, 5)},
        'Senior': {'multiplier': 1.3, 'exp_range': (5, 8)},
        'Lead': {'multiplier': 1.6, 'exp_range': (7, 12)},
        'Manager': {'multiplier': 2.0, 'exp_range': (10, 15)}
    }

    data = []
    n_anomalies = int(n_records * anomaly_rate)
    anomaly_indices = random.sample(range(n_records), n_anomalies)

    for i in range(n_records):
        # Basic employee info
        emp_id = f"TEST_{i+1:03d}"

        # Select business unit and level
        bu_name = random.choice(list(business_units.keys()))
        bu_info = business_units[bu_name]

        level = random.choice(list(levels.keys()))
        level_info = levels[level]

        # Generate consistent demographic data
        experience = random.randint(level_info['exp_range'][0], level_info['exp_range'][1])
        age = 23 + experience + random.randint(0, 3)

        # Dates
        join_date = datetime(2020, 1, 1) + timedelta(days=random.randint(0, 1000))
        pay_date = datetime(2024, 6, 15)
        tenure_months = max(1, (pay_date - join_date).days // 30)

        # Calculate NORMAL salary patterns (what autoencoder should learn)
        base_annual = np.random.normal(bu_info['base_salary'], bu_info['std'])
        base_annual = max(25000, base_annual)  # Minimum salary floor

        # Apply level multiplier
        adjusted_annual = base_annual * level_info['multiplier']
        monthly_salary = adjusted_annual / 12

        # Standard allowance and deduction patterns
        allowance = monthly_salary * bu_info['allowance_rate'] * random.uniform(0.9, 1.1)
        statutory_bonus = monthly_salary * 0.0417  # Fixed bonus rate

        # Gross salary calculation
        gross_salary = monthly_salary + allowance + statutory_bonus

        # Standard deduction patterns
        deductions = gross_salary * bu_info['deduction_rate'] * random.uniform(0.95, 1.05)
        net_salary = gross_salary - deductions

        # Additional features that autoencoders can learn patterns from
        salary_per_experience = monthly_salary / max(1, experience)
        allowance_ratio = allowance / monthly_salary
        deduction_ratio = deductions / gross_salary
        net_to_gross_ratio = net_salary / gross_salary

        # Initialize anomaly tracking
        is_anomaly = i in anomaly_indices
        anomaly_type = "Normal"
        expected_reconstruction_error = "Low"

        # INJECT ANOMALIES - Patterns that should break autoencoder reconstruction
        if is_anomaly:
            anomaly_scenarios = [
                "extreme_salary", "inverted_ratios", "inconsistent_deductions",
                "outlier_allowance", "negative_values", "data_corruption"
            ]
            scenario = random.choice(anomaly_scenarios)

            if scenario == "extreme_salary":
                # Extremely high or low salary compared to pattern
                if random.random() < 0.5:
                    monthly_salary *= random.uniform(3.0, 5.0)  # Extreme overpay
                else:
                    monthly_salary *= random.uniform(0.2, 0.4)  # Extreme underpay
                gross_salary = monthly_salary + allowance + statutory_bonus
                net_salary = gross_salary - deductions
                anomaly_type = "Extreme Salary"
                expected_reconstruction_error = "Very High"

            elif scenario == "inverted_ratios":
                # Allowance higher than salary (unusual pattern)
                allowance = monthly_salary * random.uniform(1.2, 2.0)
                gross_salary = monthly_salary + allowance + statutory_bonus
                net_salary = gross_salary - deductions
                anomaly_type = "Inverted Ratios"
                expected_reconstruction_error = "High"

            elif scenario == "inconsistent_deductions":
                # Deductions don't follow the learned pattern
                if random.random() < 0.5:
                    deductions = gross_salary * random.uniform(0.5, 0.8)  # Too high
                else:
                    deductions = gross_salary * random.uniform(0.0, 0.03)  # Too low
                net_salary = gross_salary - deductions
                anomaly_type = "Inconsistent Deductions"
                expected_reconstruction_error = "High"

            elif scenario == "outlier_allowance":
                # Allowance completely out of pattern
                allowance = monthly_salary * random.uniform(0.3, 0.6)  # Way too high
                gross_salary = monthly_salary + allowance + statutory_bonus
                net_salary = gross_salary - deductions
                anomaly_type = "Outlier Allowance"
                expected_reconstruction_error = "Medium"

            elif scenario == "negative_values":
                # Impossible negative values
                if random.random() < 0.5:
                    net_salary = -abs(net_salary)  # Negative net salary
                else:
                    deductions = gross_salary * 1.2  # Deductions exceed gross
                    net_salary = gross_salary - deductions
                anomaly_type = "Negative Values"
                expected_reconstruction_error = "Very High"

            elif scenario == "data_corruption":
                # Corrupted data patterns
                monthly_salary = monthly_salary * random.choice([0.01, 100])  # Decimal point errors
                allowance = allowance * random.choice([0.1, 10])
                gross_salary = monthly_salary + allowance + statutory_bonus
                net_salary = gross_salary - deductions
                anomaly_type = "Data Corruption"
                expected_reconstruction_error = "Very High"

            # Recalculate ratios after anomaly injection
            salary_per_experience = monthly_salary / max(1, experience)
            allowance_ratio = allowance / monthly_salary if monthly_salary != 0 else 0
            deduction_ratio = deductions / gross_salary if gross_salary != 0 else 0
            net_to_gross_ratio = net_salary / gross_salary if gross_salary != 0 else 0

        # Create comprehensive record for autoencoder testing
        record = {
            # Identifiers
            'Employee_ID': emp_id,
            'Business_Unit': bu_name,
            'Employee_Level': level,

            # Demographic features (for autoencoder input)
            'Age': age,
            'Experience_Years': experience,
            'Tenure_Months': tenure_months,

            # Salary features (main patterns for autoencoder)
            'Monthly_Salary': round(monthly_salary, 2),
            'Allowance': round(allowance, 2),
            'Statutory_Bonus': round(statutory_bonus, 2),
            'Gross_Salary': round(gross_salary, 2),
            'Total_Deductions': round(deductions, 2),
            'Net_Salary': round(net_salary, 2),

            # Derived features (pattern recognition)
            'Salary_Per_Experience': round(salary_per_experience, 2),
            'Allowance_Ratio': round(allowance_ratio, 4),
            'Deduction_Ratio': round(deduction_ratio, 4),
            'Net_to_Gross_Ratio': round(net_to_gross_ratio, 4),
            'Gross_to_Base_Ratio': round(gross_salary / monthly_salary if monthly_salary != 0 else 0, 4),
            'Total_Compensation': round(gross_salary + allowance, 2),

            # Ground truth for evaluation
            'Is_Anomaly': 1 if is_anomaly else 0,
            'Anomaly_Type': anomaly_type,
            'Expected_Reconstruction_Error': expected_reconstruction_error,

            # Additional metadata
            'Pay_Period': pay_date.strftime('%Y-%m'),
            'Join_Date': join_date.strftime('%Y-%m-%d')
        }

        data.append(record)

    return pd.DataFrame(data)

def prepare_autoencoder_features(df):
    """
    Prepare features specifically for autoencoder input
    """
    # Select numerical features that autoencoders can learn patterns from
    feature_columns = [
        'Age', 'Experience_Years', 'Tenure_Months',
        'Monthly_Salary', 'Allowance', 'Statutory_Bonus', 'Gross_Salary',
        'Total_Deductions', 'Net_Salary', 'Salary_Per_Experience',
        'Allowance_Ratio', 'Deduction_Ratio', 'Net_to_Gross_Ratio',
        'Gross_to_Base_Ratio', 'Total_Compensation'
    ]

    X = df[feature_columns].copy()

    # Handle any infinite or NaN values
    X = X.replace([np.inf, -np.inf], np.nan)
    X = X.fillna(X.median())

    # Normalize features (important for autoencoders)
    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X)

    return X_scaled, feature_columns, scaler

def evaluate_autoencoder_performance(df, reconstruction_errors, threshold=None):
    """
    Evaluate autoencoder performance using reconstruction errors
    """
    true_labels = df['Is_Anomaly'].values

    # If no threshold provided, use percentile-based threshold
    if threshold is None:
        threshold = np.percentile(reconstruction_errors, 80)  # Top 20% as anomalies

    # Convert reconstruction errors to binary predictions
    predictions = (reconstruction_errors > threshold).astype(int)

    print("=== AUTOENCODER ANOMALY DETECTION EVALUATION ===\n")

    # Display threshold information
    print(f"Reconstruction Error Threshold: {threshold:.6f}")
    print(f"Reconstruction Error Statistics:")
    print(f"  Mean: {np.mean(reconstruction_errors):.6f}")
    print(f"  Std:  {np.std(reconstruction_errors):.6f}")
    print(f"  Min:  {np.min(reconstruction_errors):.6f}")
    print(f"  Max:  {np.max(reconstruction_errors):.6f}")
    print()

    # Confusion Matrix
    cm = confusion_matrix(true_labels, predictions)
    print("Confusion Matrix:")
    print(f"                 Predicted")
    print(f"                 Normal  Anomaly")
    print(f"Actual Normal    {cm[0,0]:6d}  {cm[0,1]:7d}")
    print(f"Actual Anomaly   {cm[1,0]:6d}  {cm[1,1]:7d}")
    print()

    # Calculate metrics
    tp, tn, fp, fn = cm[1,1], cm[0,0], cm[0,1], cm[1,0]

    accuracy = (tp + tn) / (tp + tn + fp + fn)
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0

    print("Performance Metrics:")
    print(f"Accuracy:    {accuracy:.3f} ({accuracy*100:.1f}%)")
    print(f"Precision:   {precision:.3f} ({precision*100:.1f}%)")
    print(f"Recall:      {recall:.3f} ({recall*100:.1f}%)")
    print(f"F1-Score:    {f1_score:.3f}")
    print(f"Specificity: {specificity:.3f} ({specificity*100:.1f}%)")

    # ROC AUC Score
    try:
        auc_score = roc_auc_score(true_labels, reconstruction_errors)
        print(f"ROC AUC:     {auc_score:.3f}")
    except:
        print("ROC AUC:     Could not calculate")
    print()

    # Analysis by anomaly type and expected reconstruction error
    print("=== ANOMALY TYPE ANALYSIS ===")
    anomaly_df = df[df['Is_Anomaly'] == 1].copy()
    if len(anomaly_df) > 0:
        anomaly_df['Reconstruction_Error'] = [reconstruction_errors[i] for i in anomaly_df.index]
        anomaly_df['Detected'] = [predictions[i] for i in anomaly_df.index]

        type_analysis = anomaly_df.groupby(['Anomaly_Type', 'Expected_Reconstruction_Error']).agg({
            'Detected': ['count', 'sum'],
            'Reconstruction_Error': ['mean', 'max']
        }).round(6)

        type_analysis.columns = ['Count', 'Detected', 'Avg_Error', 'Max_Error']
        type_analysis['Detection_Rate'] = (type_analysis['Detected'] / type_analysis['Count'] * 100).round(1)

        print(type_analysis)

    # Reconstruction error distribution
    print(f"\n=== RECONSTRUCTION ERROR DISTRIBUTION ===")
    normal_errors = reconstruction_errors[true_labels == 0]
    anomaly_errors = reconstruction_errors[true_labels == 1]

    print(f"Normal Records - Reconstruction Error:")
    print(f"  Mean: {np.mean(normal_errors):.6f}")
    print(f"  Std:  {np.std(normal_errors):.6f}")
    print(f"  Max:  {np.max(normal_errors):.6f}")

    print(f"Anomaly Records - Reconstruction Error:")
    print(f"  Mean: {np.mean(anomaly_errors):.6f}")
    print(f"  Std:  {np.std(anomaly_errors):.6f}")
    print(f"  Min:  {np.min(anomaly_errors):.6f}")

    # Model performance assessment
    print(f"\n=== AUTOENCODER PERFORMANCE ASSESSMENT ===")

    error_separation = np.mean(anomaly_errors) / np.mean(normal_errors) if np.mean(normal_errors) > 0 else 0
    print(f"Error Separation Ratio: {error_separation:.2f}")

    if recall >= 0.8 and precision >= 0.7 and error_separation >= 2.0:
        print("âœ… AUTOENCODER PERFORMANCE: EXCELLENT")
        print("   - High detection rate with good precision")
        print("   - Clear separation between normal and anomaly errors")
    elif recall >= 0.6 and precision >= 0.5 and error_separation >= 1.5:
        print("âš ï¸  AUTOENCODER PERFORMANCE: GOOD")
        print("   - Decent detection with moderate precision")
        print("   - Consider adjusting threshold or model architecture")
    else:
        print("âŒ AUTOENCODER PERFORMANCE: NEEDS IMPROVEMENT")
        print("   - Low detection rate or poor precision")
        print("   - Model may need retraining or different architecture")

    # Recommendations
    print(f"\n=== RECOMMENDATIONS ===")
    if error_separation < 1.5:
        print("- Reconstruction errors are not well separated")
        print("- Consider adding more layers or changing architecture")
        print("- Try different loss functions (MAE vs MSE)")

    if precision < 0.6:
        print("- High false positive rate")
        print("- Consider increasing threshold or improving feature engineering")

    if recall < 0.6:
        print("- Missing many true anomalies")
        print("- Consider decreasing threshold or adding more training data")

    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1_score,
        'auc_score': auc_score if 'auc_score' in locals() else None,
        'threshold': threshold,
        'error_separation': error_separation,
        'reconstruction_errors': reconstruction_errors,
        'predictions': predictions,
        'true_labels': true_labels
    }

# Generate test dataset
print("Generating autoencoder test dataset...")
test_df = generate_autoencoder_test_data(n_records=50, anomaly_rate=0.20)

# Prepare features for autoencoder
X_test, feature_names, scaler = prepare_autoencoder_features(test_df)

print(f"\n=== TEST DATASET SUMMARY ===")
print(f"Total Test Records: {len(test_df)}")
print(f"Normal Records: {len(test_df[test_df['Is_Anomaly'] == 0])}")
print(f"Anomalous Records: {len(test_df[test_df['Is_Anomaly'] == 1])}")
print(f"Features for Autoencoder: {len(feature_names)}")

print(f"\nAnomaly Types in Test Set:")
print(test_df['Anomaly_Type'].value_counts())

print(f"\nExpected Reconstruction Error Levels:")
print(test_df['Expected_Reconstruction_Error'].value_counts())

# Display sample data
print(f"\n=== SAMPLE TEST DATA ===")
print(test_df[['Employee_ID', 'Monthly_Salary', 'Net_Salary', 'Allowance_Ratio',
              'Is_Anomaly', 'Anomaly_Type']].head(10))

# Simulate autoencoder reconstruction errors for demonstration
print(f"\n=== SIMULATING AUTOENCODER TESTING ===")
print("Generating simulated reconstruction errors...")

# Create realistic reconstruction errors
np.random.seed(42)
reconstruction_errors = np.zeros(len(test_df))

for i, row in test_df.iterrows():
    if row['Is_Anomaly'] == 0:
        # Normal records should have low reconstruction error
        reconstruction_errors[i] = np.random.exponential(0.01)
    else:
        # Anomalous records should have higher reconstruction error
        severity_multiplier = {
            'Low': 2.0,
            'Medium': 5.0,
            'High': 10.0,
            'Very High': 20.0
        }
        multiplier = severity_multiplier.get(row['Expected_Reconstruction_Error'], 5.0)
        reconstruction_errors[i] = np.random.exponential(0.01 * multiplier)

# Evaluate performance
results = evaluate_autoencoder_performance(test_df, reconstruction_errors)

# Save test dataset
print(f"\nSaving test dataset to 'autoencoder_test_dataset.xlsx'...")
with pd.ExcelWriter('autoencoder_test_dataset.xlsx', engine='openpyxl') as writer:
    # Test data
    test_df.to_excel(writer, sheet_name='Test_Data', index=False)

    # Features for autoencoder
    features_df = pd.DataFrame(X_test, columns=feature_names)
    features_df['Is_Anomaly'] = test_df['Is_Anomaly'].values
    features_df.to_excel(writer, sheet_name='Normalized_Features', index=False)

    # Results summary
    summary_df = pd.DataFrame({
        'Metric': ['Total Records', 'Normal Records', 'Anomalous Records', 'Anomaly Rate (%)',
                  'Simulated Accuracy', 'Simulated Precision', 'Simulated Recall', 'Simulated F1'],
        'Value': [len(test_df), len(test_df[test_df['Is_Anomaly'] == 0]),
                 len(test_df[test_df['Is_Anomaly'] == 1]),
                 round(len(test_df[test_df['Is_Anomaly'] == 1]) / len(test_df) * 100, 1),
                 round(results['accuracy'], 3), round(results['precision'], 3),
                 round(results['recall'], 3), round(results['f1_score'], 3)]
    })
    summary_df.to_excel(writer, sheet_name='Performance_Summary', index=False)

print("âœ… Test dataset generated successfully!")
print(f"\n=== HOW TO USE WITH YOUR AUTOENCODER ===")
print("1. Load the 'Normalized_Features' sheet as your test input")
print("2. Pass the features through your trained autoencoder")
print("3. Calculate reconstruction errors for each record")
print("4. Use the evaluation function above with your actual errors")
print("5. Compare against the ground truth 'Is_Anomaly' column")
print("\nExample code:")
print("```python")
print("# Load your trained autoencoder model")
print("# model = load_model('your_autoencoder.h5')")
print("# ")
print("# Get predictions and calculate reconstruction errors")
print("# reconstructed = model.predict(X_test)")
print("# reconstruction_errors = np.mean(np.square(X_test - reconstructed), axis=1)")
print("# ")
print("# Evaluate performance")
print("# results = evaluate_autoencoder_performance(test_df, reconstruction_errors)")
print("```")
````
